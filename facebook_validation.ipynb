# Age Bracket Accuracies
## CS and Pred

```python
bracket_dat.index=range(len(bracket_dat))
sbarc = bracket_dat.drop([0,3,4,7,8])
sbarc.drop(sbarc.columns[range(1,16)],axis=1,inplace=True)
[[brackets[i],sbarc.iloc[i*2,i+1]] for i in range(6)]
```

```python
brac_acc = []
for i in range(6):
    brac_acc.append(sbarc.iloc[i*2,i+1]*ratio[i]+sbarc.iloc[i*2+1,i+1]*(1-ratio[i]))
[[brackets[i],sbarc.iloc[i*2+1,i+1]] for i in range(6)]
```

## TPR

```python
acc_mat=sbarc.drop('filename',axis=1)
acc_mat=acc_mat.values.tolist()
wt_ac, ps =[[0]*6 for _ in range(12)], [cs_p, pred_p]
for i in range(12):
    for j in range(6):
        wt_ac[i][j]=acc_mat[i][j]*ps[i%2][j]
TPR=[]
for i in range(6):
    TPR.append((wt_ac[i*2][i]+wt_ac[i*2+1][i])/sum(x[i] for x in wt_ac))
[[brackets[i],TPR[i]] for i in range(6)]
```

```python
tc = [sum(sum(x) for x in wt_ac[(i*2):(i*2+2)]) for i in range(6)]
avg_tpr=sum(tc[i]*TPR[i] for i in range(6))
avg_tpr
```
